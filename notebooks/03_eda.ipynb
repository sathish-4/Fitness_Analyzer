```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

df = pd.read_csv('../data/merged_data.csv')

print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
df.head()
```

```python
print("Total users: ",df['Id'].nunique())
print("Date range: ",df['ActivityDate'].min(),df['ActivityDate'].max())
print("Total range: ",len(df))
```

```python
plt.figure(figsize=(12,5)) #we create canva
plt.plot(df['ActivityDate'], df['TotalSteps'], color='green') #draw the chart
plt.title('Daily Steps Over Time')#title
plt.xlabel('Date')#x label
plt.ylabel('Total Steps')#ylabel
plt.tight_layout()#fix spacing automatically
plt.show()#used to display the chart
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,x=df['DayOfWeek'],y=df['TotalSteps'],color='steelblue')
plt.title('Avg steps')
plt.xlabel('DayOfWeek')
plt.ylabel('TotalSteps')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x=df['TotalSteps'],y=df['Calories'],color='blue',alpha=0.5)
plt.title('Steps VS Calories')
plt.xlabel('TotalSteps')
plt.ylabel('Calories')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x=df['TotalMinutesAsleep'],y=df['VeryActiveMinutes'],color=('red'),alpha=(0.5))
plt.title('Sleep vs Active Minutes')
plt.xlabel('Steep(minutes)')
plt.ylabel('Active minutes')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df , x='Id',y='TotalSteps',color='yellow')
plt.title('Most Active Users')
plt.xlabel('id (user ID)')
plt.ylabel('TotalSteps')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
activity_data = df[['VeryActiveMinutes','FairlyActiveMinutes',
                    'LightlyActiveMinutes','SedentaryMinutes']].mean()

plt.bar(activity_data.index, activity_data.values, 
        color=['green','blue','orange','red'])
plt.title('Average Minutes by Activity Type')
plt.xlabel('Activity Type')
plt.ylabel('Average Minutes')
plt.tight_layout()
plt.show()
```
